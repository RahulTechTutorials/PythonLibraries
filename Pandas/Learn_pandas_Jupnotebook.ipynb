```python
##Using map function in pandas(getting another column using join)

df = pd.DataFrame()
df['id'] = [1,2,3,4,5]
df['names'] = ['Rick','Anthony','Bill','Jen','Jill']
df['city'] = ['Las Vegas','New York','Albany','Orlando','Milford']
df
ref_dict = {'Las Vegas':'Nevada',
            'New York':'New York',
            'Albany':'New York',
            'Orlando':'Florida',
            'Milford':'Connecticut'
           }

df['state'] = df['city'].map(ref_dict)

###Or we can do it with another dataframe , converting it into dictionary
df2 = pd.DataFrame()
df2['city'] = ['Las Vegas','New York','Albany','Orlando','Milford']
df2['state'] = ['Nevada','New York','New York','Florida','Connecticut']
df2
city_to_state_dict = dict(zip(df2['city'],df2['state']))
city_to_state_dict
df['state'] = df['city'].map(city_to_state_dict)
df
```

```python
###String manipulation functions
import pandas as pd
import os
os.getcwd()
os.listdir()
df = pd.read_csv('train.csv')
df
df2 = pd.DataFrame(columns = ['col1'])
df2['col1'] = df.iloc[1:10,3]
##to convert the string in lower
df2['col1'].str.lower()
##to convert the string in upper
df2['col1'].str.upper()
##to convert the string in Capitalize
df2['col1'].str.capitalize()
##to convert the string in Title
df2['col1'].str.title()
##to split the string
df2,df2['col1'].str.split(' ')
##to find the string endwith
df2,df2['col1'].str.endswith('ard')
```
