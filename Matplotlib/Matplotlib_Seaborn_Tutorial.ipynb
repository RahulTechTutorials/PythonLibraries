```python
######Linear graphs #####

####Getting the simplest plot done

from matplotlib import pyplot as plt
plt.plot([1,2,3,4],[2,5,6,7])
plt.show()
```

```python
######Linear graphs #####
###Now lets generate the X,y,lables,titles

X=[1,2,3,4]
y=[2,5,6,7]

plt.plot(X,y)
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
######Linear graphs #####
###now lets add style to your plots for that you import matplotlib.style

from matplotlib import style

style.use('ggplot') ##We will use ggplot

X=[1,2,3,4]
y=[2,5,6,7]


X2= [12,13,16,19]
y2 = [1,3,6,12]

plt.plot(X,y,'g',label='line1',linewidth=3) ##g signifies green color, line1 is the name and linewidth is the widhth
plt.plot(X2,y2,'r',label='line2',linewidth=3)
plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend() ##to display the legends
plt.grid(True,color='w') ##to display the grid lines

plt.show()
```

```python
###Bar graphs ######
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot') ##We will use ggplot

X=[1,2,3,4]
y=[2,5,6,7]


X2= [12,13,16,19]
y2 = [1,3,6,12]

plt.bar(X,y,label='bar1') ###just use bar instead of line
plt.bar(X2,y2,label='bar2')
plt.title('Bar Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend() 

plt.show()
```

```python
###Histogram plot ######
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot') ##We will use ggplot

X=[1,2,3,4,12,34,55,64,23,89]
bins=[2,5,6,7,11,23,45,60,70,90]
plt.hist(X,bins,histtype='bar',) ###just use hist instead of bar and specify histtype=bar

plt.title('Histogram Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
###Scatter plot ######
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot') ##We will use ggplot

X=[1,2,3,4,12,34,55,64,23,89]
y=[2,5,6,7,11,23,45,60,70,90]
plt.scatter(X,y) ###just use hist instead of bar and specify histtype=bar

plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
### box plot####
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure(figsize=[10,6])
X=[1,2,3,4,12,34,55,64,23,30,113]
Y = [3,4,5,6,32,12,56,98,22,37]
plt.subplot(121)
plt.boxplot(X,showmeans=True,meanline = True)
plt.title('X box plot')
plt.ylabel('X')
plt.subplot(122)
plt.boxplot(Y,showmeans=True,meanline = True)
plt.title('Y box plot')
plt.ylabel('Y')
```

# SEABORN - WITH SUBPLOTS
Now lets look at some advanced visualisation with seaborn
We will use subplots with axis
First we create the grid and then assign values of the axis to each plot

```python
import seaborn as sns
import os
##Lets look at seaborn now
os.getcwd()
os.listdir('/Users/rahuljain/Desktop/Python/PythonLibraries/Matplotlib')
import pandas as pd
df = pd.read_csv('train.csv')
fig, axis = plt.subplots(2, 3,figsize=(14,10))

sns.barplot(x='Pclass',y='Survived',data=df,ax=axis[0,0],hue='Sex',palette='spring')
sns.barplot(x=df['Embarked'],y='Survived',data=df,ax=axis[0,1],hue='Sex',palette='winter')
sns.barplot(x=df['Embarked'],y='Survived',data=df,ax=axis[0,2],palette='spring')
sns.barplot(x=df['Cabin'].str.slice(0,1),y='Survived',data=df,ax=axis[1,0],palette='spring')
sns.barplot(x=df['Name'].str.split(",",expand=True)[1].str.split(".",expand=True)[0],y='Survived',data=df,ax=axis[1,1],palette='winter')
sns.barplot(x=pd.qcut(df['Fare'],4,['Low','Medium','High','Very_high']),y='Survived',data=df,ax=axis[1,2],hue='Sex',palette='spring')
```

# FACET GRID - Broken up graphs by row,col,hue
We will first create the Facet Grid, pass the dataframe, will tell seaborn on what basis we want to break the subplots dowwn, row,col,hue
and then map the sns plots on top of it

```python
fg = sns.FacetGrid(df,col="Pclass")
fg.map(sns.pointplot,'Embarked', 'Survived', 'Sex', palette='deep')
fg.add_legend()
```

```python
fg = sns.FacetGrid(df,col="Embarked")
fg.map(sns.pointplot,'Pclass', 'Survived', 'Sex', palette='deep')
fg.add_legend()
```

```python
a = sns.FacetGrid( df, hue = 'Survived', aspect=4 )
a.map(sns.kdeplot, 'Age', shade= True )
##a.set(xlim=(0 , df['Age'].max()))
a.add_legend()
```

```python
his = sns.FacetGrid(df,row='Sex',col = 'Pclass',hue='Survived')
his.map(plt.hist,'Age',alpha=0.80)
his.add_legend()
```

```python
his = sns.FacetGrid(df,row='Sex',col = 'Embarked',hue='Survived')
his.map(plt.hist,'Age',alpha=0.80)
his.add_legend()
```

```python
import warnings
warnings.filterwarnings("ignore")
pp = sns.pairplot(df[['Pclass','Fare','Age','Survived']], hue = 'Survived', palette = 'deep', size=4.2, diag_kind = 'kde')
```
