# Group by

````python
import pandas as pd

# Define the data as a list of dictionaries
data = [
    {"date": "1/1/2017", "city": "new york", "temperature": 32, "windspeed": 6, "event": "Rain"},
    {"date": "1/2/2017", "city": "new york", "temperature": 36, "windspeed": 7, "event": "Sunny"},
    {"date": "1/3/2017", "city": "new york", "temperature": 28, "windspeed": 5, "event": "Snow"},
    {"date": "1/4/2017", "city": "new york", "temperature": 33, "windspeed": 7, "event": "Sunny"},
    {"date": "1/1/2017", "city": "mumbai", "temperature": 90, "windspeed": 5, "event": "Sunny"},
    {"date": "1/2/2017", "city": "mumbai", "temperature": 85, "windspeed": 12, "event": "Fog"},
    {"date": "1/3/2017", "city": "mumbai", "temperature": 87, "windspeed": 15, "event": "Fog"},
    {"date": "1/4/2017", "city": "mumbai", "temperature": 92, "windspeed": 5, "event": "Rain"},
    {"date": "1/1/2017", "city": "paris", "temperature": 45, "windspeed": 10, "event": "Sunny"},
    {"date": "1/2/2017", "city": "paris", "temperature": 50, "windspeed": 13, "event": "Cloudy"},
    {"date": "1/3/2017", "city": "paris", "temperature": 54, "windspeed": 8, "event": "Cloudy"},
    {"date": "1/4/2017", "city": "paris", "temperature": 42, "windspeed": 10, "event": "Cloudy"}
]

# Create DataFrame
df = pd.DataFrame(data)

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Display the DataFrame
print(df)
````

````python
df
````

````python
df.date = pd.to_datetime(df.date)
````

# Concat

````python
import pandas as pd
india_weather = {
    "city" : ["mumbai","delhi","bangalore"],
    "temperature" : [32,45,30],
    "humidity" : [80,60,78]
}

us_weather = {
    "city" : ["new york","chicago","orlando"],
    "temperature" : [21,14,35],
    "humidity" : [68,65,75]
}
````

````python
df_india = pd.DataFrame(india_weather)
df_us = pd.DataFrame(us_weather)
````

````python
df_india
````

````python
pd.concat([df_india, df_us], ignore_index=True)
````

````python
new_df = pd.concat([df_india, df_us], keys=['india','us'])
````

````python
new_df
````

````python
new_df.loc['india']
````

````python
temperature_df = pd.DataFrame({
    "city" : ["delhi","mumbai","bangalore"],
    "temperature" : [32,45,30]
},index=[2,1,3])
````

````python
temperature_df
````

````python
windspeed_df = pd.DataFrame({
    "city" : ["mumbai","delhi","bangalore"],
    "windspeed" : [7,12,9]
}, index=[1,2,3])
````

````python
windspeed_df
````

````python
pd.concat([temperature_df,windspeed_df],axis = 1)
````

````python
##add a series to the dataframe

s = pd.Series(['Humid','Dry','Rain'], name='event',index=[1,2,3])
````

````python
new_df = pd.concat([temperature_df,s], axis=1)
````

````python
new_df
````
