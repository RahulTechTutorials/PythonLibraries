```
def main():
    """This program is to print the pyramid"""
    shape = int(input("Enter the type of shape. 1 for triangle and 2 for right pyramid and 3 for inverted pyramid"))
    assert (shape == 1 or shape == 2 or shape ==3)
    nrows = int(input("Enter the number of rows for the  pyramid/triangle, odd number fo pyramid please"))
    assert nrows > 0
    if shape == 1:
        printtriange(nrows)
    elif shape ==2:
        printpyramid(nrows)
    else:
        printinvertedpyramid(nrows)

def printtriange(n):
    for i in range(1,n+1):
        print('*' * i)

def printpyramid(n):
    for i in range(1,n+1):
        print(' ' * (n-i) , '*' * (2*i-1))

def printinvertedpyramid(n):
    for i in range(1,n+1):
        print(' ' * (i-1) , '*' * ((2*(n-i)) + 1))

if __name__ == "__main__":
    main()
```

# Dictionary comprehension - builds a dictionary from an iterable


```
products = ["laptop", "phone", "tablet"]
prices = [999, 699, 499]

# Build a price lookup dictionary from two lists
price_lookup = { product:price for product,price in zip(products,prices)}
print(price_lookup)
```

```
# Filter: only products over $600
new_product_list = { product:price for product,price in zip(products,prices) if price>600 }
new_product_list
```

# Shallow copy vs Deep Copy

```
from copy import copy,deepcopy
original = [[0,1],[2,3],[4,5]]
shallow = copy(original)
deep = deepcopy(original)
original[0][0] = 99
print(f'original[0][0] : {original[0][0]}')
print(f'shallow[0][0] : {shallow[0][0]}')
print(f'deep[0][0] : {deep[0][0]}')
```

# Exception handling - try, except, else, and finally blocks to handle errors gracefully.

```
def load_sales_data(filepath):
    try:
        with open(filepath, "r") as f:
            data = f.read()
    except FileNotFoundError:
        print(f"File not found: {filepath}")
        data = None
    except PermissionError:
        print("Permission denied.")
        data = None
    else:
        print("File loaded successfully.")
    finally:
        print("Attempted file load.")
    return data
```

# Python decorator - *A decorator is a function that wraps another function to extend its behavior without modifying it directly. They’re applied using the @ syntax.*

```
import time

def timer(func):
  def wrapper(*args, **kwargs):
    start = time.time()
    output = func(*args, **kwargs)
    end = time.time()
    print(f'{func} function ran in {end - start:.4f}s')
    return output
  return wrapper

@timer
def square3(x):
  return x ** 2
square3(7)
```

# Python generator - *A generator produces items one at a time, on demand, rather than loading everything into memory at once. You create one with the yield keyword or a generator expression.*

```
m_list = [x for x in range(1000)] ## list
m_list_gen = (x for x in range(1000)) ##generator
print(type(m_list), type(m_list_gen))
```

```
def read_in_chunks(filepath, chunksize=1000):
  with open(filepath,'r') as f:
    chunk = []
    for line in f():
      chunk.append(line)
      if len(chunk) == chunksize:
        yield chunk
        chunk=[]
  if chunk:
    yield chunk
```

```
m_list_gen = (x for x in range(1000)) ##generator
for i in m_list_gen:
  if i > 10:
    break
  print(i)
```

# GIl - Global Interpreterer Lock - No true parallelism. Workaround

```
from multiprocessing import pool

def process_chunk(data_chunk):
  return [x**2 for x in data_chunk]


data = list(range(16))
chunks = [ data[x:x+4] for x in range(0,16,4)]
with Pool(4) as p:
  result = p.map(process_chunk,chunks)
print (result)
```
